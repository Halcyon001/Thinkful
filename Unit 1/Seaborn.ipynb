```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import date
from datetime import timedelta
```

```python
#Load data from csv into dataframe
#kickstarter data
ksdata = pd.read_csv('intro course/capstone/kickstarter_projects/ksprojects.csv')

print(ksdata.columns)
```

Time to plot the prices of each room, in 4 different types of plots with seaborn, a boxplot, a bar plot, pointplot, and histogram

```python
#boxplot
#sns.set(style='ticks')
sns.set(style='whitegrid')
fig, ax = plt.subplots() 
fig.set_size_inches(11.7, 8.27) 

boxplot = sns.boxplot(x=ksdata.main_category, y=ksdata.usd_goal_real, data=ksdata, palette='pastel')
boxplot.set(ylabel='Project Goal (USD)', xlabel='Main Project Category')
plt.title('Project Goals by Category')
plt.ylim(0, 51000)
plt.xticks(rotation=45)
sns.despine(trim=True)

plt.show()
```

```python
#barplot
sns.set(style='whitegrid')
fig, ax = plt.subplots() 
fig.set_size_inches(11.7, 8.27)

barplot = sns.barplot(x=ksdata.main_category, y=ksdata.usd_goal_real, data=ksdata, palette='pastel', ci=95)
barplot.set(ylabel='Project Goal USD', xlabel='Main Project Category')
fig.set_size_inches(11.7, 8.27) 
plt.title('Project Goals by Category')
plt.ylim(0, 150000)
plt.xticks(rotation=45)
sns.despine(trim=True)

plt.show()
```

```python
#pointplot
sns.set(style='darkgrid')
fig, ax = plt.subplots() 
fig.set_size_inches(11.7, 8.27)

pointplot = sns.pointplot(x=ksdata.main_category, y=ksdata.usd_goal_real, data=ksdata, palette='pastel', ci=95)
pointplot.set(ylabel='Project Goal USD', xlabel='Main Project Category')
fig.set_size_inches(11.7, 8.27) 
plt.title('Project Goals by Category')
plt.ylim(0, 150000)
plt.xticks(rotation=45)
sns.despine(trim=True)

plt.show()
```

```python
#histogram
sns.set(style='whitegrid')
fig.set_size_inches(11.7, 8.27)

plt.hist(ksdata.usd_goal_real, range=(0,25000)) 
plt.title('Project Goals')
plt.xticks(rotation=45)
sns.despine(trim=True)

plt.show()
```
