```python
dummies = pd.get_dummies(y2015.drop('loan_status', 1))
```

```python
corrs = dummies.corr().abs()
#iterate over one of the triangles (upper or lower) by label and drop one of whichever variable has a high correlation
#doesn't matter which one
#select one of the triangles
print(dummies.shape)
triupper = corrs.where(np.triu(np.ones(corrs.shape), k=1).astype(np.bool))

highly_corred_vars = [col for col in triupper.columns if any(triupper[col] > 0.90)]

dummies = dummies.drop(highly_corred_vars, axis=1)

print(dummies.shape)
```

```python
from sklearn import ensemble
from sklearn.model_selection import cross_val_score

rfc = ensemble.RandomForestClassifier()
#X = y2015.drop('loan_status', 1)
Y = y2015['loan_status']
X = dummies
X = X.dropna(axis=1)
```

```python
print(cross_val_score(rfc, X, Y, cv=10))
print(np.mean(cross_val_score(rfc, X, Y, cv=10)))
```

```python
rfc.fit(X, Y)
```

```python
features = list(zip(X, rfc.feature_importances_))
features.sort(key=lambda x: x[1])
```

```python
from sklearn.feature_selection import SelectKBest

print(X.shape)
kselector = SelectKBest(k=10)
X_best = kselector.fit_transform(X, Y)
```

```python
print(X_best.shape)
scores = cross_val_score(rfc, X_best, Y, cv=10)
print(scores)
print(np.mean(scores))
```
