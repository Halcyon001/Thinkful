```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def printvitals(sample):
    print('Mean of sample: ' + str(sample.mean()))
    print('Standard deviation of sample: ' + str(sample.std()))
```

```python
pop1 = np.random.binomial(10, .2, 10000)
pop2 = np.random.binomial(10, .5, 10000)

plt.hist(pop1, alpha=.5, label='Population 1')
plt.hist(pop2, alpha=.5, label='Population 2')
plt.legend(loc='upper right')

plt.show()

printvitals(pop1)
printvitals(pop2)
```

```python
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

print('Sample 1:')
printvitals(sample1)
print('Sample 2:')
printvitals(sample2)

plt.hist(sample1, alpha=.5, label='sample 1')
plt.hist(sample2, alpha=.5, label='sample 2')
plt.legend(loc='upper right')
plt.show()
```
